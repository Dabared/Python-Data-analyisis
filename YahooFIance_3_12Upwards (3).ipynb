# Accessing Fiancial data available in Yahoo Finance

Herein, we demonstrate how to access financial data from Yahoo Finance using Python. We can download company financial information available in Yahoo Finance, as well as historical market data by using python interfaces.

### Version dependencies

Python older versions (before 3.12), access is through the Pandas data-reader. However, newer versions we use directly the Yahoo-Finance interface: yfinance.

### Time Series Data
A time series is a sequence of numerical data points taken at successive equally spaced points in time. In investing, a time series tracks the movement of the chosen data points, such as the stock price, over a specified period of time with data points recorded at regular intervals. All the financial data available as time-series data.

### Multiple tickers
To initialize multiple Ticker objects, use

```python
mtickers = yf.Tickers('msft aapl goog')

# access each ticker using (example)
price_msft = mtickers.tickers['MSFT'].history(period="1mo")
price_aapl = mtickers.tickers['AAPL'].history(period="3mo")
price_goog = mtickers.tickers['GOOG'].history(period="3mo")
```

```python
type(price_msft)
```

```python
mtickers
```

```python
price_msft.head()
```

```python
comTable = pd.DataFrame()
```

```python
comTable["AAPL_Open"] = price_aapl.Open
comTable["MSFT_Open"] = price_msft.Open
comTable["GOOG_Open"] = price_goog.Open
```

```python
comTable.plot()
```

### Financial Plots

mplfinance is one of the package that suppor for stock market illustrations
- After importing mplfinance, plotting OHLC data is as simple as calling mpf.plot() on the dataframe

```python
import mplfinance as mpf
```

```python
price_msft
```

```python
#price_msft
mpf.plot(price_msft)
```

```python
mpf.plot(price_msft, type='candle')
```

```python
price_msft_2y = mtickers.tickers['MSFT'].history(period="2y")
h = price_msft_2y.Open.plot()
```

```python
type(h)
plt.savefig('destination_path.eps', format='eps')
```

```python
mpf.plot(price_msft_2y,type='renko')
```

You can also plot moving averages with the mav keyword

use a scalar for a single moving average
use a tuple or list of integers for multiple moving averages

```python
mpf.plot(price_msft,type='ohlc',mav=3)
```

```python
mpf.plot(price_msft,type='candle',mav=(3,6,9))
```
