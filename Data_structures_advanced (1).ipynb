### Dictionaries
Another useful data type built into Python is the dictionary. Dictionaries are sometimes found in other languages as associative memories or associative arrays. Unlike sequences, which are indexed by a range of numbers, dictionaries are indexed by keys, which can be any immutable type; strings and numbers can always be keys.

It is best to think of a dictionary as a set of key: value pairs, with the requirement that the keys are unique (within one dictionary). A pair of braces creates an empty dictionary: {}. Placing a comma-separated list of key:value pairs within the braces adds initial key:value pairs to the dictionary; this is also the way dictionaries are written on output.



#### Recall the dictionery operations

1. Create a dictionary containing account information
    - Tyeps of the account

```python
accounts = {'tyep':{'Saving', 'Current'}, 'bank': {"BOC","SAM","COM"}, 'year': 2000}
print(accounts)
print("Keys", accounts.keys())
print("Items", accounts.items())
for key in accounts.keys():
    print("the key = ", key, "has items", accounts[key])

for key, value in accounts.items():
    print("the key = ", key, "has items", value, "of the type", type(value))

for key in accounts:
    print(key)
```

```python
def update_accounts(accounts, key, value):
    if key in accounts:
        accounts[key].add(value)
    else:
        accounts[key] = set()
        accounts[key].add(value)
```

```python
names = list(data.Name)
```

```python
name_set = set(names)
```

```python
print(name_set)
print("Number of distinct names =", len(name_set))
```

```python
first_name = list()
last_name = list()
for i in range(len(names)):
    name_parts = names[i].split()
    first_name.append(name_parts[0])
    last_name.append(name_parts[1])

print("First names  = ", set(first_name))
print("Last names = ", set(last_name))
```

```python
first_name_set = set()
last_name_set = set()
for name in name_set:
    name_parts = name.split()
    first_name_set.add(name_parts[0])
    last_name_set.add(name_parts[1])

print("First names = ", first_name_set)
print("Last names = ", last_name_set)
```

```python
# What most frequent contact type used by the people who have the surname "Williams"?

# We use a dictionery to structure the contact types. The idea is to get names associated with different types of contacts.
# dict(contact_type:names)

contacts = dict()

```

```python
def update_contacts(contacts, contact_type, name):
    if contact_type in contacts:
        contacts[contact_type].append(name)
    else:
        contacts[contact_type] = list()
        contacts[contact_type].append(name)

```

```python
for i in range(len(data.contact)):
    update_contacts(contacts, data.contact[i], last_name[i])
```

```python
def get_count(name, name_list):
    i = 0
    for i in range(len(name_list)):
        if name == name_list[i]:
            i = i+1
    return i
```

```python
get_count("Williams", contacts['cellular'])
```

```python
get_count("Williams", contacts['unknown'])
```

```python
get_count("Williams", contacts['telephone'])
```

```python
def update_list_dict(list_dict, key, value):
    if key in list_dict:
        list_dict[key].append(value)
    else:
        list_dict[key] = list()
        list_dict[key].append(value)

```

```python
williams_frequency = dict()
for contact_type in contacts.keys():
    update_list_dict(williams_frequency, contact_type, get_count("Williams", contacts[contact_type]))

```
